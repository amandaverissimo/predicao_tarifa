####Dados Tarifas Aéreas Internacionais
Foram coletados 20 arquivos a serem explorados. Os dados estão segmentados por escopo temporal (mensal) - cada um dos documentos corresponde aos registros de tarifas aéreas comercializadas no Brasil ao público adulto em  geral, inclusive as vendas pela internet, independentemente do país em que apágina encontra-se hospedada. Exclui-se aquelas adquiridas com descontos restritos a grupos específicos, programas de milhagem, entre outras condições. A análise terá como lapso temporal 01/2023 a 08/2024.<br>
<br>
Cada um dos arquivos exibe os seguintes dados:<br>
ANO: Indica o ano em que as passagens registradas foram vendidas, independentemente da data do voo. <br>
MES: Indica o mês em que as passagens registradas foram vendidas, independentemente da data do voo. <br>
EMPRESA: Sigla de 3 letras estabelecida pela Organização de Aviação Civil Internacional que indica a empresa que registrou os dados. <br>
ORIGEM: Sigla de 4 letras que indica os aeródromos de origem no Brasil. <br>
DESTINO: Sigla de 4 letras que indica os aeródromos de destino no exterior. <br>
RETORNO: Sigla de 4 letras que indica os aeródromos de retorno do passageiro, independentemente de escalas ou conexões. <br>
CLASSE IDA: Indica a classe de cabine da passagem vendida, com os códigos Y (para classe econômica), J (para classe executiva) e F (para primeira classe). <br>
CLASSE VOLTA: Indica a classe de cabine da passagem vendida, com os códigos Y (para classe econômica), J (para classe executiva), F (para primeira classe) ou 9 (usado para preencher o campo de classe de volta indicando passagens de apenas ida). <br>
TARIFA: Apresenta o valor do serviço de transporte aéreo constante da passagem. Não inclui as tarifas de embarque dos aeroportos nem valores cobrados por serviços opcionais. Para as tarifas aéreas internacionais, é apresentado o valor nominal em Dólares americanos (USD).<br>
ASSENTOS: número de assentos vendidos no mês de referência para a chave composta por ano e mês de venda da passagem, empresa aérea, aeroportos de origem, destino e retorno, classes de cabine de ida e volta e valor de tarifa.  <br>
<br>

#####Obtendo a distância entre ORIGEM e DESTINO:

```python
# Configurar a API Nominatim
geolocator = Nominatim(user_agent="airfare_prediction", timeout=10)

def get_coordinates(icao_code):
    try:
        location = geolocator.geocode(icao_code)  # Utilizando sigla ICAO
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None

# Função para calcular a distância haversine
def haversine(lon1, lat1, lon2, lat2):
    R = 6371.0  # Raio da Terra em quilômetros
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

# Carregar o dataset
data_flight = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Datasets MVP Machine Learning & Analytics/no_outliers_data_flight.csv')

# Obter as siglas únicas
unique_icaos = pd.concat([data_flight['ORIGEM'], data_flight['DESTINO'], data_flight['RETORNO']])
unique_icaos = unique_icaos[unique_icaos != '9999'].unique()

# Criar um dicionário para armazenar as coordenadas
coordinates_dict = {icao: get_coordinates(icao) for icao in unique_icaos}

# Aplicar as coordenadas ao dataset
data_flight[['LAT_ORIGEM', 'LON_ORIGEM']] = data_flight['ORIGEM'].apply(lambda code: pd.Series(coordinates_dict.get(code, (None, None))))
data_flight[['LAT_DESTINO', 'LON_DESTINO']] = data_flight['DESTINO'].apply(lambda code: pd.Series(coordinates_dict.get(code, (None, None))))
data_flight[['LAT_RETORNO', 'LON_RETORNO']] = data_flight.apply(lambda row: pd.Series(coordinates_dict.get(row['RETORNO'], (None, None))) if row['RETORNO'] != '9999' else pd.Series([None, None]), axis=1)

# Calcular a distância para cada linha do dataset (ida e volta)
data_flight['DISTANCIA_IDA'] = data_flight.apply(lambda row: haversine(row['LON_ORIGEM'], row['LAT_ORIGEM'],
                                                                      row['LON_DESTINO'], row['LAT_DESTINO']), axis=1)
data_flight['DISTANCIA_VOLTA'] = data_flight.apply(lambda row: haversine(row['LON_DESTINO'], row['LAT_DESTINO'],
                                                                       row['LON_RETORNO'], row['LAT_RETORNO']) if row['RETORNO'] != '9999' else 0, axis=1)

# Remover as colunas de latitude e longitude
data_flight = data_flight.drop(columns=['LAT_ORIGEM', 'LON_ORIGEM', 'LAT_DESTINO', 'LON_DESTINO', 'LAT_RETORNO', 'LON_RETORNO'])

# Exibir o dataframe final com apenas as colunas de distância
print(data_flight[['ORIGEM', 'DESTINO', 'RETORNO', 'DISTANCIA_IDA', 'DISTANCIA_VOLTA']])
```

####Treinamento do modelo



O modelo utilizado para treinamento da máquina preditiva foi o Random Forest. Foi empregada validação cruzada através do k-fold, com divisão reduzida para 3, levando em conta a densidade dos dados. A função RandomizedSearch realizou a busca aleatória com 10 iterações dos melhores hiperparâmetros para ajuste e assertividade do algoritmo.<br>
<br>
Anteriormente, havia tentado empregar o GridSearch por crer que a assertividade seria maior. No entanto, há um alto custo computacional associado ao GridSearch e, após 3 horas de execução, a atividade da célula foi abortada, optando-se por usar o RandomizedSearch como alternativa.<br>
<br>
Levando em conta a quantidade de entradas e visualização da "big picture", é pertinente desconsiderar o GridSearch. Embora o GridSearch avalie muitos pontos no espaço de hiperparâmetros, é provável que o espectro de busca seja menor. Por outro lado, o RandomizedSearch pode percorrer um espectro maior, avaliando alguns pontos, o que aumenta as chances de capturar melhores resultados. Esta abordagem equilibra a necessidade de eficiência computacional com a probabilidade de encontrar uma configuração de hiperparâmetros otimizada.

```python
# Configurando a validação cruzada K-Fold no conjunto de treino
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Lista de hiperparâmetros para Random Forest
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, None],
    'min_samples_split': [2, 10],
    'min_samples_leaf': [1, 2]
}

# Configurando o RandomizedSearchCV com validação cruzada
randomized_search = RandomizedSearchCV(RandomForestRegressor(), param_distributions=param_grid, n_iter=10, cv=kf, scoring='neg_mean_squared_error', n_jobs=-1)

# Treinando o modelo com RandomizedSearchCV
randomized_search.fit(X_train, y_train)

# Obtenção dos melhores hiperparâmetros
best_params = randomized_search.best_params_
print(f'Melhores parâmetros: {best_params}')

# Prevendo os resultados no conjunto de validação e teste final
y_val_pred = randomized_search.predict(X_val)
y_test_pred = randomized_search.predict(X_test)

# Avaliando o desempenho do modelo
val_mse = mean_squared_error(y_val, y_val_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print(f'MSE no conjunto de validação: {val_mse:.2f}')
print(f'MSE no conjunto de teste final: {test_mse:.2f}')
```

Inicialmente os hiperparâmetros do modelo foram determinados pelo buscador Random Search cross-validation (RandomizedSearchCV), que busca os melhores hiperparâmetros avaliando a performance do modelo em combinações aleatórias de valores de hiperparêmetro. A melhor combinação de hiperparâmetros que produz o melhor desempenho em um conjunto de validação é selecionada.<br>
<br>
<br>

Na tentativa de otimizar o desempenho do modelo foi executada a validação cruzada repetida, porém o resultado foi bastante próximo do já obtido. Pelo Desvio Padrão vê-se que a variação dos MSE entre as diferentes repetições e divisões é relativamente pequena.


```python
# Configurar o Repeated K-Fold
rkf = RepeatedKFold(n_splits=5, n_repeats=3, random_state=42)

# Criando uma instância do modelo RandomForestRegressor
model = RandomForestRegressor() # Criando o modelo

# Realizar a validação cruzada repetida
cv_scores = cross_val_score(model, X_train, y_train, cv=rkf, scoring='neg_mean_squared_error')

# Calcular a média e o desvio padrão dos scores de MSE
mean_mse_rkf = -cv_scores.mean()
std_mse_rkf = cv_scores.std()

print(f'Média do MSE com Repeated K-Fold: {mean_mse_rkf:.2f}')
print(f'Desvio padrão do MSE com Repeated K-Fold: {std_mse_rkf:.2f}')
```

Para melhor parametrização do MSE dos conjuntos de teste e validação, é interessante ter como referência o valor da baseline. A linha de base é uma métrica simples que costuma ser a média dos valores da variável dependente - no caso, TARIFA_USD. Nesse cenário, a média serve como uma previsão constante para todos os casos.<br>
Se o MSE do modelo for menor do que o MSE da baseline, isso é um sinal positivo. Indica que o modelo está oferecendo previsões mais precisas e eficazes do que a abordagem simples de usar a média como previsão.
Essa comparação simples ajuda a determinar se o uso de um modelo mais complexo como Random Forest realmente vale a pena e é vantajoso para a predição

```python
# Calcular a média dos valores de treino
mean_baseline = np.mean(y_train)

# Prever os valores de validação e teste com a média
y_val_pred_baseline = np.full_like(y_val, mean_baseline)
y_test_pred_baseline = np.full_like(y_test, mean_baseline)

# Calcular o MSE para o baseline
val_mse_baseline = mean_squared_error(y_val, y_val_pred_baseline)
test_mse_baseline = mean_squared_error(y_test, y_test_pred_baseline)

print(f'MSE no conjunto de validação (Baseline - Média): {val_mse_baseline:.2f}')
print(f'MSE no conjunto de teste final (Baseline - Média): {test_mse_baseline:.2f}')
```

```python
# Definição da grade de hiperparâmetros para Gradient Boosting
param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 10],
    'min_samples_leaf': [1, 2]
}

# Configurando o RandomizedSearchCV com validação cruzada
randomized_search_gb = RandomizedSearchCV(GradientBoostingRegressor(), param_distributions=param_grid_gb, n_iter=10, cv=kf, scoring='neg_mean_squared_error', n_jobs=-1)

# Treinando o modelo com RandomizedSearchCV
randomized_search_gb.fit(X_train, y_train)

# Obtenção dos melhores hiperparâmetros
best_params_gb = randomized_search_gb.best_params_
print(f'Melhores parâmetros para Gradient Boosting: {best_params_gb}')

# Prevendo os resultados no conjunto de validação e teste final
y_val_pred_gb = randomized_search_gb.predict(X_val)
y_test_pred_gb = randomized_search_gb.predict(X_test)

# Avaliando o desempenho do modelo
val_mse_gb = mean_squared_error(y_val, y_val_pred_gb)
test_mse_gb = mean_squared_error(y_test, y_test_pred_gb)

print(f'MSE no conjunto de validação (Gradient Boosting): {val_mse_gb:.2f}')
print(f'MSE no conjunto de teste final (Gradient Boosting): {test_mse_gb:.2f}')
```

```python
# Definir os modelos base
estimators = [
    ('rf', RandomForestRegressor(random_state=42)),
    ('gb', GradientBoostingRegressor(random_state=42))
]

# Definir o modelo de stacking
stacking_model = StackingRegressor(
    estimators=estimators,
    final_estimator=LinearRegression(),
    cv=5
)

# Treinar o modelo de stacking no conjunto de treino
stacking_model.fit(X_train, y_train)

# Prever os resultados no conjunto de validação
y_val_pred = stacking_model.predict(X_val)

# Avaliar o desempenho do modelo no conjunto de validação
val_mse_st = mean_squared_error(y_val, y_val_pred)
print(f'MSE no conjunto de validação (Stacking): {val_mse_st:.2f}')

# Prever os resultados no conjunto de teste
y_test_pred = stacking_model.predict(X_test)

# Avaliar o desempenho do modelo no conjunto de teste
test_mse_st = mean_squared_error(y_test, y_test_pred)
print(f'MSE no conjunto de teste final (Stacking): {test_mse_st:.2f}')
```

O uso de ensemble com o Linear Regressor não se mostrou uma solução efetiva para esse caso, já que apresentou os mesmos resultados obtidos com o Random Forest. É possível que o RandomForest esteja dominando a previsão no ensemble por favorecimento do Linear Regressor ou que o Gradient Boosting não contribua com novas informações para o processo, além de poder estar capturando padrões muito próximos aos do Randon Forest.
Para melhor avaliação o processo foi realizado com o Ridge Regression, que apresentou igual performance:

```python
# Utilizando Ridge Regression como meta-modelo
stacking_model_ridge = StackingRegressor(
    estimators=estimators,
    final_estimator=Ridge(),
    cv=5
)

stacking_model_ridge.fit(X_train, y_train)
y_val_pred_ridge = stacking_model_ridge.predict(X_val)
val_mse_ridge = mean_squared_error(y_val, y_val_pred_ridge)
y_test_pred_ridge = stacking_model_ridge.predict(X_test)
test_mse_ridge = mean_squared_error(y_test, y_test_pred_ridge)
print(f'MSE no conjunto de validação (Stacking com Ridge): {val_mse_ridge:.2f}')
print(f'MSE no conjunto de teste final (Stacking com Ridge): {test_mse_ridge:.2f}')
```

A acurácia corresponde a proporção de previsões corretas feitas pelo modelo no conjunto de teste, para a previsão em questão a acurácia foi de aproximadamente 95%.<br>
Já para a predição das tarifas aéreas, um problema de regressão, foi utilizado o MSE (Mean Squared Error), que penaliza mais fortemente erros maiores. O MSE corresponde a média dos quadrados dos erros. Por não ser tão intuitivo e de fácil entendimento / mensuração como a acurácia, o resultado do melhor modelo empenhado foi convertido para margem de erro. Dessa forma o parâmetro se torna mais legível.<br>
